```python
from gtts import gTTS #for creating audio file 
import os 
import IPython
import IPython.display
import tempfile
```

```python
text_file=input("Please enter a string: ")
print(text_file)

audio_file=gTTS(text=text_file,lang="en",slow=False)

with tempfile.NamedTemporaryFile(suffix=".mp3", delete=False) as temp_audio:
    audio_file.write_to_fp(temp_audio)
    temp_audio_path = temp_audio.name
    
IPython.display.Audio(temp_audio_path)    
    
```
